## Import Libraries

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

# Adjusts the configuration of the plots we will create 
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
```

## Read in the data

```python
df = pd.read_csv(r'C:\Users\16606\Downloads\archive (15)\movies.csv')
```

## Looking at data

```python
df.head()
```

## Looking for any missing data

```python
df = df.dropna(how='any',axis=0) 
```

```python
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

## Data types for columns

```python
df.dtypes
```

## Changing data type of votes, budget, gross columns to remove the decimal and zero

```python
df['votes'] = df['votes'].astype('int64')

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')
```

## Creating corrected year column

```python
df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)

df.head()
```

## Sorting data by gross revenue ascending

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows', None)
```

## Drop any duplicates

```python
df['company'].drop_duplicates().sort_values(ascending=False)
```

## Building Scatterplot Budget vs Gross Revenue

```python
plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Budget for Film')

plt.ylabel('Gross Earnings')

plt.show()
```

```python
df.head()
```

## Plot the Budget vs Gross using Seaborn

```python
sns.regplot(x='budget', y='gross', data = df, scatter_kws = {"color": "red"}, line_kws = {"color": "blue"})
```

## Looking at correlation

```python
df.corr()
```

```python
correlation_matrix = df.corr()

sns.heatmap(correlation_matrix, annot = True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()
```

```python
correlation_matrix = df_numerized.corr()

sns.heatmap(correlation_matrix, annot = True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs.head()
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs.head()
```

```python
high_corr = sorted_pairs[(sorted_pairs) > .5]

high_corr.head()
```
